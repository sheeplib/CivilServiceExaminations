## Step4：拆分url分离可变参数

```python
# 分离page之前的url串
try:
    page_pos = url.find('page=')
    page_pos += 5
    url_before_page = url[:page_pos]
    url_after_page = url[page_pos:-1]
    print(url_before_page)
except:
    print('找不到关键字‘page=’')
# 分离page之后的url串    
try:
    after_page_pos = url_after_page.find('&')
    url_after_page = url_after_page[after_page_pos:-1]
    print(url_after_page)
except:
    print('找不到关键字‘&’')
    
```

成功分离后合法url可以由url_before_page+str(page)+url_after_page构成  
page = 1,2,3...

```python
page = 1
url= url_before_page + str(page) + url_after_page
print(url)
```
